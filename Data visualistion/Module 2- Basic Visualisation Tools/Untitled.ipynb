```python
df_canada['Total'] = df_canada.sum(axis=1)
# let us view the first five elements and see how the dataframe was changed
df_canada.head()
```

```python
years = list(map(str, range(1980, 2014)))
```

```python
df_canada.set_index('Country', inplace=True)

# let us view the first five elements and see how the dataframe was changed
df_canada.head()
```

```python
#I will use the inline backend
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df_canada.sort_values(['Total'],axis = 0 ,ascending = False, inplace = True)
```

```python
df_top = df_canada.head()
```

```python
df_top
```

```python
df_top = df_top[years].transpose()
df_top
```

```python
# let us change the index values of df_top5 to type integer for plotting
df_top.index = df_top.index.map(int) 
df_top.plot(kind='area', 
             alpha=0.25, # 0-1, default value a= 0.5
             stacked=True,
             figsize=(20, 10), # pass a tuple (x, y) size
             )

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
# get the 5 countries with the least contribution
df_least5 = df_canada.tail(5)

 # transpose the dataframe
df_least5 = df_least5[years].transpose() 
df_least5.head()
df_least5.index = df_least5.index.map(int) # let's change the index values of df_least5 to type integer for plotting
df_least5.plot(kind='area',stacked=False, alpha=0.45, figsize=(20, 10)) 
plt.title('Immigration Trend of 5 Countries with Least Contribution to Immigration')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```
