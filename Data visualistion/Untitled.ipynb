```python
data_path = r'Police_Department_Incidents_-_Previous_Year__2016_.csv'
incident_df = pd.read_csv(data_path)
```

```python
incident_df.head()
```

```python
incident_df = incident_df.head(100)
# df_incident = df_incident.iloc[:100,:]
incident_df.tail()
```

```python
# San Francisco latitude and longitude values
latitude = 37.77
longitude = -122.42
# create map and display it
sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=8)

# display the map of San Francisco
sanfran_map
```

```python
 #instantiate a feature group for the incidents in the dataframe
incidents = folium.map.FeatureGroup()
for long, lat in zip(incident_df.X, incident_df.Y):
    # loop through the 100 crimes and add each to the incidents feature group
    incidents.add_child(
        folium.CircleMarker(
            [lat, long],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

# add incidents to map
sanfran_map.add_child(incidents)
sanfran_map
```
