```python
!wget --quiet https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/alice_novel.txt

# open the file and read it into a variable alice_novel
alice_novel = open('alice_novel.txt', 'r').read()
    
print ('File downloaded and saved!')
```

```python
!pip install wordcloud
```

```python
alice_novel
```

```python
stopwords = set(STOPWORDS)
# instantiate a word cloud object
alice_wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

# generate the word cloud
alice_wc.generate(alice_novel)
```

```python
# display the word cloud
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,12))
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
alice_wc = WordCloud(
    background_color='white',
    max_words=200,
    stopwords=stopwords
)

import re
parkinson =  re.sub('[^a-zA-Z]', ' ', alice_novel)
parkinson = parkinson.lower()
parkinson = parkinson.split(' ')
# Steamming
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
parkinson = [ps.stem(word) for word in parkinson]
parkinson = ' '.join(parkinson)
# generate the word cloud
alice_wc.generate(parkinson)
```

```python
fig = plt.figure(figsize=(10,12))
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import numpy as np
from PIL import Image
!wget --quiet https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Images/alice_mask.png

alice_mask = np.array(Image.open('alice_mask.png'))
```

```python
fig = plt.figure(figsize=(12,12))
plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# instantiate a word cloud object
alice_wc = WordCloud(background_color='white', max_words=500, mask=alice_mask, stopwords=stopwords)

# generate the word cloud
alice_wc.generate(alice_novel)

# display the word cloud
fig = plt.figure(figsize=(12,12))
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
